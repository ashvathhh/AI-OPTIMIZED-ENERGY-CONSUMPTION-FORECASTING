```python
# Let's first load the dataset, take a look at the first few rows to understand the structure, 
# and then perform data visualizations to extract insights.

import pandas as pd

# Load the dataset
file_path = 'household_power_consumption_with_current.csv'
data = pd.read_csv(file_path)

# Show the first few rows of the dataset to understand its structure
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'], format="%d/%m/%Y")
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True, infer_datetime_format=True)
```

```python
import matplotlib.pyplot as plt

# Convert 'Date' and 'Time' columns to datetime format
data['DateTime'] = pd.to_datetime(data['Date'] + ' ' + data['Time'], format='%d-%m-%Y %H:%M:%S')

# Set 'DateTime' as the index for easier time-series plotting
data.set_index('DateTime', inplace=True)

# Visualize trends for Global Active Power, Reactive Power, and Voltage
plt.figure(figsize=(14, 8))

# Plot Global Active Power
plt.subplot(3, 1, 1)
plt.plot(data.index, data['Global_active_power'], color='blue', label='Global Active Power (W)')
plt.title('Global Active Power over Time')
plt.ylabel('Power (W)')
plt.grid(True)

# Plot Global Reactive Power
plt.subplot(3, 1, 2)
plt.plot(data.index, data['Global_reactive_power'], color='orange', label='Global Reactive Power (W)')
plt.title('Global Reactive Power over Time')
plt.ylabel('Power (W)')
plt.grid(True)

# Plot Voltage
plt.subplot(3, 1, 3)
plt.plot(data.index, data['Voltage'], color='green', label='Voltage (V)')
plt.title('Voltage over Time')
plt.ylabel('Voltage (V)')
plt.grid(True)

plt.tight_layout()
plt.show()

# Distribution of Current Consumption
plt.figure(figsize=(10, 6))
plt.hist(data['current_consumption'], bins=50, color='purple')
plt.title('Distribution of Current Consumption')
plt.xlabel('Current Consumption (A)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Hourly trends of Current Consumption
data['Hour'] = data.index.hour

plt.figure(figsize=(10, 6))
plt.plot(data.groupby('Hour')['current_consumption'].mean(), marker='o', linestyle='-', color='red')
plt.title('Average Current Consumption by Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Current Consumption (A)')
plt.grid(True)
plt.show()
```

```python
# Group the data by hours and calculate the average current consumption for each hour
hourly_avg_consumption = data.groupby('Hour')['current_consumption'].mean()

# Find the hours with the highest current consumption
peak_hours = hourly_avg_consumption.nlargest(5)

# Plot the hourly current consumption, highlighting peak hours
plt.figure(figsize=(10, 6))
plt.plot(hourly_avg_consumption.index, hourly_avg_consumption, color='blue', label='Average Current Consumption', marker='o')

# Highlight the peak hours
plt.scatter(peak_hours.index, peak_hours, color='red', label='Peak Hours', s=100, zorder=5)

# Label the plot
plt.title('Average Hourly Current Consumption with Peak Hours Highlighted')
plt.xlabel('Hour of the Day')
plt.ylabel('Current Consumption (A)')
plt.grid(True)
plt.legend()
plt.xticks(range(0, 24))
plt.show()

# Display the peak hours and their consumption values
peak_hours
```

From the above graph, we can see that the 5 top hours are the most current consumed hours.
